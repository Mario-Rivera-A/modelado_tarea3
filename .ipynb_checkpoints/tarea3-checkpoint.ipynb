```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Inicializamos la cantidad de agentes y la formaci√≥n deseada
n_agents = 50  # M√≠nimo de 50 agentes
np.random.seed(42)  # Para reproducibilidad de la simulaci√≥n

# Generamos posiciones iniciales aleatorias para los agentes en el plano XY
positions = np.random.rand(n_agents, 2) * 10  # Espacio inicial de 10x10 para los agentes

# Definimos la formaci√≥n deseada: un c√≠rculo de radio 5 centrado en (5, 5)
circle_center = np.array([5, 5])
circle_radius = 5
angles = np.linspace(0, 2 * np.pi, n_agents, endpoint=False)
desired_positions = circle_center + circle_radius * np.column_stack((np.cos(angles), np.sin(angles)))

# Definimos los par√°metros de control
alpha = 0.05  # Factor de adaptaci√≥n de velocidad (cuanto m√°s r√°pido se mueven los agentes)
threshold = 0.1  # Umbral para considerar que el agente ha alcanzado su posici√≥n deseada

# Creamos la figura para la animaci√≥n
fig, ax = plt.subplots()
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
scat = ax.scatter(positions[:, 0], positions[:, 1], color='blue')
ax.set_title('Simulaci√≥n de Sistema Multi-Agente en Formaci√≥n Circular')

# Definimos la funci√≥n que actualiza la posici√≥n de los agentes en cada cuadro de la animaci√≥n
def update(frame):
    global positions
    # Calculamos la diferencia entre la posici√≥n deseada y la actual para cada agente
    diff = desired_positions - positions
    # Calculamos la nueva posici√≥n sumando una fracci√≥n (alpha) de la diferencia a la posici√≥n actual
    positions += alpha * diff
    # Actualizamos el gr√°fico de dispersi√≥n con las nuevas posiciones
    scat.set_offsets(positions)
    return scat,

# Creamos la animaci√≥n
anim = FuncAnimation(fig, update, frames=500, interval=50, blit=True)

```
