```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Inicializamos la cantidad de agentes y la formación deseada
n_agents = 50  # Mínimo de 50 agentes
np.random.seed(42)  # Para reproducibilidad de la simulación

# Generamos posiciones iniciales aleatorias para los agentes en el plano XY
positions = np.random.rand(n_agents, 2) * 10  # Espacio inicial de 10x10 para los agentes

# Definimos la formación deseada: un círculo de radio 5 centrado en (5, 5)
circle_center = np.array([5, 5])
circle_radius = 5
angles = np.linspace(0, 2 * np.pi, n_agents, endpoint=False)
desired_positions = circle_center + circle_radius * np.column_stack((np.cos(angles), np.sin(angles)))

# Definimos los parámetros de control
alpha = 0.05  # Factor de adaptación de velocidad (cuanto más rápido se mueven los agentes)
threshold = 0.1  # Umbral para considerar que el agente ha alcanzado su posición deseada

# Creamos la figura para la animación
fig, ax = plt.subplots()
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
scat = ax.scatter(positions[:, 0], positions[:, 1], color='blue')
ax.set_title('Simulación de Sistema Multi-Agente en Formación Circular')

# Definimos la función que actualiza la posición de los agentes en cada cuadro de la animación
def update(frame):
    global positions
    # Calculamos la diferencia entre la posición deseada y la actual para cada agente
    diff = desired_positions - positions
    # Calculamos la nueva posición sumando una fracción (alpha) de la diferencia a la posición actual
    positions += alpha * diff
    # Actualizamos el gráfico de dispersión con las nuevas posiciones
    scat.set_offsets(positions)
    return scat,

# Creamos la animación
anim = FuncAnimation(fig, update, frames=500, interval=50, blit=True)

```
